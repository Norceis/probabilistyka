```python
import matplotlib.pyplot as plt
import scipy as scp
import numpy as np

fig, ax = plt.subplots(1, 1)

x = np.arange(-1, 1.1, 0.01)

scales = [0.2, 0.5, 0.8, 1.1, 1.4, 1.7, 2.0]
colors = ['red', 'green', 'blue', 'magenta', 'brown', 'black', 'yellow', 'orange']

for i in range(len(scales)):
    ax.plot(x, scp.stats.norm.pdf(x, scale=scales[i]), color=colors[i],
            lw=5, alpha=0.9, label=f'std {scales[i]}')

plt.xlabel('Zmienna')
plt.ylabel('Prawdopodobieństwo')
ax.legend()
plt.rcParams["figure.figsize"] = (6,6)
plt.show()
```

```python
fig, ax = plt.subplots(1, 1)

for i in range(len(scales)):
    ax.plot(x, scp.stats.norm.cdf(x, scale=scales[i]), color=colors[i],
            lw=5, alpha=0.9, label=f'std {scales[i]}')

plt.xlabel('Zmienna')
plt.ylabel('Skumulowane prawdopodobieństwo')
ax.legend()
plt.show()
```

```python
fig, (ax1, ax2) = plt.subplots(2, 1, constrained_layout=False)

for i in range(len(scales)):
    ax1.plot(x, scp.stats.norm.pdf(x, scale=scales[i]), color=colors[i],
            lw=5, alpha=0.9, label=f'std {scales[i]}')

for i in range(len(scales)):
    ax2.plot(x, scp.stats.norm.cdf(x, scale=scales[i]), color=colors[i],
            lw=5, alpha=0.9, label=f'std {scales[i]}')

ax1.set_xlabel('Zmienna')
ax1.set_ylabel('Prawdopodobieństwo')
ax1.legend()
ax2.set_xlabel('Zmienna')
ax2.set_ylabel('Skumulowane prawdopodobieństwo')
ax2.legend()
plt.show()
```

```python
fig, ax = plt.subplots(1, 1)

x = np.arange(0, 10, 0.05)

ax.scatter(x, scp.stats.binom.pmf(x, 10, 0.5), color='red', label='Bi(10, 0.5)', marker='.')
ax.scatter(x, scp.stats.binom.pmf(x, 10, 0.1), color='green', label='Bi(10, 0.1)', marker='+')
plt.xlabel('Zmienna')
plt.ylabel('Prawdopodobieństwo')
ax.legend()
plt.show()
```

```python
fig, ax = plt.subplots(1, 1)

x = np.arange(0, 20, 0.5)

ax.scatter(x, scp.stats.poisson.pmf(x, 5), color='magenta', label='Rozkład Poissona', marker='+')
ax.vlines(x, 0, scp.stats.poisson.pmf(x,5), colors='orange', lw=1, alpha=0.5)
plt.xlabel('Zmienna')
plt.ylabel('Prawdopodobieństwo')
ax.legend()
plt.show()
```
