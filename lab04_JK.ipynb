```python
import matplotlib.pyplot as plt
from functools import reduce

conditions = [(0.01, 0.8),(0.5, 0.98)]

class Burglar():
    def __init__(self, p_I, conditions):
        self.p_I = p_I
        self.conditions = conditions
        self.events = 1
        self.events_if = 1
        for x in range(len(conditions)):
            self.events *= (conditions[x][0])
            self.events_if *= (conditions[x][1])
        
    def p_I_if_conditions(self):
        return self.p_I * self.events_if / self.events
    
        
test = Burglar(0.002, conditions)
outcome = test.p_I_if_conditions()
print(outcome)
```

```python
conditions = [(0.1, 0.1), (0.02, 0.2), (0.3, 0.93)]
test = Burglar(0.005, conditions)
outcome = test.p_I_if_conditions()
print(outcome)
```

```python
conditions = [(0.1, 0.6), (0.02, 0.2), (0.3, 0.93)]

plt.xlabel('Different samples')
plt.ylabel('Probability of burglary if conditions were met')
plt.scatter(1, Burglar(0.005, conditions[:1]).p_I_if_conditions(), color='g', label='(0.1, 0.6), (0.02, 0.2)')
plt.scatter(2, Burglar(0.005, conditions[1:2]).p_I_if_conditions(), color='b', label='(0.02, 0.2), (0.3, 0.93)')
plt.scatter(3, Burglar(0.005, conditions[0:2]).p_I_if_conditions(), color='y', label='(0.1, 0.6), (0.02, 0.2), (0.3, 0.93)')
plt.legend()
plt.show()
```
