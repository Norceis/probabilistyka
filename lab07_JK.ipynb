```python
import arviz as az
import pymc3 as pm
import numpy as np

def the_thing(yes_count):
    with pm.Model() as model:
        p_d = pm.Uniform("p_d", 0, 1)
        p_observed = pm.Deterministic("p_observed", 0.25 + 0.5*p_d)
        answers = pm.Binomial("answers", 100, p_observed, observed=yes_count)
        idata = pm.sample(2000, tune=2500, return_inferencedata=True)
    
    az.plot_trace(idata, show=True)
```

```python
the_thing(70)
```

```python
the_thing(20)
```

```python
the_thing(10)
```

```python
the_thing(5)
```

Wynik eksperymentu ukazuje kilka obserwacji: prawdopodobieństwo otrzymania odpowiedzi twierdzącej na tak zadane pytanie w ankiecie mieści się zawsze w zakresie 25 - 75%, ponieważ jest ono zależne od stanu faktycznego danego studenta. Jeśli 0% studentów zażywa narkotyki, to odpowiedź w ankiecie wyniesie w przybliżeniu 25%, jeśli jest to 100% to będzie to 75%. Dlatego taki eksperyment ma sens, ponieważ zależność procenta populacji studentów zażywających narkotyki jest w sposób liniowy powiązana z odpowiedziami twierdzącymi w ankiecie - wyniki tej ankiety należy przedstawić przy zmienionym zakresie na osi Y, wtedy otrzymamy prawdziwy procent studentów zażywających narkotyki.
